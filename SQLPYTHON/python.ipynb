```python
import pandas as pd
import numpy as np
from IPython.display import display as dp
```

Pregunta: ¿Cuáles son los 5 productos más vendidos (por cantidad) en el último mes de la tabla ventas?



```python
df=pd.read_csv('ventas.csv')
```

```python
dp(df)
```

```python
df['fecha']=pd.to_datetime(df['fecha'])
```

```python
fc=df['fecha'].max()-pd.Timedelta(days=30)
```

```python
print(fc)
```

```python
dff=df[df['fecha'] >= fc]
```

```python
dp(dff)
```

```python
top5=dff.groupby('producto')['cantidad'].sum().reset_index().sort_values(by='cantidad', ascending= False).head(5).reset_index(drop='index')
```

```python
dp(top5)
```

```python
top5=df.groupby('cliente')['cantidad'].sum().reset_index().sort_values(by='cantidad', ascending= False).head(5).reset_index(drop='index')
```

```python
dp(top5)
```
